```python
#compute watch count
def feature_count(p): 
    return p.count(',')
```

```python
#calculating count for columns

train_data['title_count'] = ht_dt['titles'].map(feature_count)
train_data['genres_count'] = ht_dt['genres'].map(feature_count)
train_data['cities_count'] = ht_dt['cities'].map(feature_count)
train_data['tod_count'] = ht_dt['tod'].map(feature_count)
train_data['dow_count'] = ht_dt['dow'].map(feature_count)
```

```python
ht_dt['g1'] = [re.sub(pattern='\:\d+',repl='',string=x) for x in ht_dt['genres']]
ht_dt['g1'] = ht_dt['g1'].apply(lambda x: x.split(','))
t1 = pd.Series(ht_dt['g1']).apply(frozenset).to_frame(name='t_genre')

for t_genre in frozenset.union(*t1.t_genre):
    t1[t_genre] = t1.apply(lambda _: int(t_genre in _.t_genre), axis=1)
```

```python
ht_dt['g4'] = [re.sub(pattern='\:\d+',repl='',string=x) for x in ht_dt['cities']]
ht_dt['g4'] = ht_dt['g4'].apply(lambda x: x.split(','))
t4 = pd.Series(ht_dt['g4']).apply(frozenset).to_frame(name='t_cities')

for t_cities in frozenset.union(*t4.t_cities):
    t4[t_cities] = t4.apply(lambda _: int(t_cities in _.t_cities), axis=1)
```
